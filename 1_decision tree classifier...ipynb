```python
import numpy as np 
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
```

```python
data.feature_names
```

```python
data.target
```

```python
# create dataframe
df = pd.DataFrame(np.c_[data.data,data.target],columns=[list(data.feature_names)+['target']])
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
# split data
x = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=2020)
```

```python
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
# train the decision tree model
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier = DecisionTreeClassifier(criterion="gini")
```

```python
classifier.fit(xtrain,ytrain)
```

```python
classifier.score(xtest,ytest)
```

```python
classifier_entropy = DecisionTreeClassifier(criterion="entropy")
classifier_entropy.fit(xtrain,ytrain)
```

```python
classifier_entropy.score(xtest,ytest)
```

```python
# feature scalling 
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler()
```

```python
sc.fit(xtrain)
```

```python
xtrain_sc = sc.transform(xtrain)
xtest_sc = sc.transform(xtest)
```

```python
classifier_sc = DecisionTreeClassifier(criterion='gini')
classifier_sc.fit(xtrain_sc,ytrain)
classifier_sc.fit(xtest_sc,ytest)
```

```python
classifier_sc.score(xtest_sc,ytest)
```
